Topic modeling & evaluation (for bad reviews):

Preparation for topic modeling:

```python
bow_vectorizer_reviews = CountVectorizer()
bow_bad_reviews = bow_vectorizer_reviews.fit_transform(normalized_bad_reviews)
bow_feature_names_bad_reviews = bow_vectorizer_reviews.get_feature_names()
```

Preparation for model evaluation:

```python
bad_reviews_tokenized = [tokenize_text(normalized_bad_reviews[doc_id]) for doc_id in range(len(normalized_bad_reviews))]
bad_reviews_dictionary = Dictionary(bad_reviews_tokenized)
bad_reviews_bow = [bad_reviews_dictionary.doc2bow(doc) for doc in bad_reviews_tokenized] #bow=bag of words
```

```python
def get_topic_words(vectorizer, lda_model, n_words):
    keywords = np.array(vectorizer.get_feature_names())
    topic_words = []
    for topic_weights in lda_model.components_:
        top_word_locs = (-topic_weights).argsort()[:n_words]
        topic_words.append(keywords.take(top_word_locs).tolist())
    return topic_words
```

Topic modeling & evaluation:

```python
cmscore=[]
ll=[]
pp=[]
lda_bad_reviews_list=[]
for no_topics_bad_reviews in range(3,13):
    #modeling
    lda_bad_reviews = LatentDirichletAllocation(n_components=no_topics_bad_reviews, max_iter=100,random_state = 42).fit(bow_bad_reviews)
    lda_bad_reviews_list.append(lda_bad_reviews)
    #evaluation
    topic_keywords = get_topic_words(vectorizer = bow_vectorizer_reviews, lda_model = lda_bad_reviews, n_words=20)
    cm = CoherenceModel(topics=topic_keywords, corpus = bad_reviews_bow, dictionary = bad_reviews_dictionary, coherence='u_mass')
    cmscore.append(np.round(cm.get_coherence(),4))
    ll.append(lda_bad_reviews.score(bow_bad_reviews))
    pp.append(lda_bad_reviews.perplexity(bow_bad_reviews))
```

```python
results=[cmscore,ll,pp]
pd.DataFrame(results, 
             columns = ["no_topics_" + str(i) for i in range(3,13)],
             index = ["cmscore (higher, better)","ll (higher, better)","pp(lower, better)"]) 
```

Visualization for best model:

```python
pyLDAvis.enable_notebook()
visualization_panel = pyLDAvis.sklearn.prepare(lda_bad_reviews_list[1], bow_bad_reviews, bow_vectorizer_reviews, mds='tsne')
visualization_panel
```
